This week I'm going to do some data exploration. This dataset is titled "West Los Angeles Traffic Collisions By Year", reflecting traffic collision incidents in the City of Los Angeles from 1/1/2022 to 12/31/2023.

```python
# import python libraries
import pandas as pd
import geopandas as gpd
```

```python
# read and add shapefile to notebook
collision = gpd.read_file('Data/Traffic_Collision_Data_from_2022_to_2023.csv')
collision
```

```python
# save it as a variable
area_count = collision['Area Name'].value_counts()
area_count
```

```python
# determine data type of area_count
type(area_count)
```

```python
# convert the series into a dataframe
area_count = area_count.reset_index()
area_count
```

```python
# determine data type of area_count
type(area_count)
```

```python
# rename columns
area_count.columns = ['Area Name', 'Number of Collision incidents']
area_count
```

```python
# draw a chart
area_count.plot()
```

```python
# draw a bar chart
area_count.plot.bar(x = 'Area Name', y = 'Number of Collision incidents', title = 'Number of Traffic Collision Incidents in Each Area')
```
