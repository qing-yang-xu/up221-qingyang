This notebook is about community comparison. Based on private vehicle traffic accidents and public transit accidents, we generated a choropleth map. We choose two community, downtown LA and Pacific palisades. Downtown LA represented high collision rate community, Pacific palisades represented low collision rate community.

```python
# importing python libraries
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
# convert FIPS, STATE, and COUNTY to string
Vehicle = pd.read_csv(
    'Group_Project/Group_Assignment_2/Qingyang_Group_Assignment_2/data/R13562228_SL140.csv',
    dtype=
    {
        'Geo_FIPS':str,
        'Geo_STATE':str,
        'Geo_COUNTY': str
    }
)
Vehicle.head()
```

```python
Vehicle = Vehicle.dropna(axis=1,how="all")
columns_to_keep = ['Geo_FIPS',
                   'SE_A10054B_001',
                   'SE_A10054B_002',
                   'SE_A10054B_003',
                   'SE_A10054B_004',
                   'SE_A10054B_005',
                   'SE_A10054B_006',
                   'SE_A10054B_007']
Vehicle2 = Vehicle[columns_to_keep]
Vehicle2.columns = ['FIPS',
'Renter-Occupied_Housing_Units',
'No_Vehicle_Available',
'1_Vehicle_Available',
'2_Vehicle_Available',
'3_Vehicle_Available',
'4_Vehicle_Available',
'5_or_More_Vehicles_Available']
Vehicle2.sample(5)
```

```python
tracts=gpd.read_file('Group_Project/Group_Assignment_2/Qingyang_Group_Assignment_2/data/Census_Tracts_2020.geojson')
tracts.head()
```

```python
tracts = tracts[['CT20','geometry']]
tracts['FIPS'] ='06' + '037' + tracts['CT20']
tracts['FIPS'] = tracts['FIPS'].astype(int)
Vehicle2['FIPS'] = Vehicle2['FIPS'].astype(int)

tracts_vehicle=tracts.merge(Vehicle2,on="FIPS")
tracts_vehicle.head()
```

```python
# Create a new column "3 or More Vehicles Available"
tracts_vehicle['3_or_More_Vehicles_Available'] = tracts_vehicle['3_Vehicle_Available'] + tracts_vehicle['4_Vehicle_Available'] + tracts_vehicle['5_or_More_Vehicles_Available']
tracts_vehicle.head()
```

# Comparison of Proportions

## Calculate the percentages

```python
# calculate PACIFIC PALISADES percentage

x1 = sum(df_sorted['No_Vehicle_Available'].tolist())   #sum of No_Vehicle
x2 = sum(df_sorted['1_Vehicle_Available'].tolist())   #sum of 1_Vehicle
x3 = sum(df_sorted['2_Vehicle_Available'].tolist())   #sum of 2_Vehicle
x4 = sum(df_sorted['3_or_More_Vehicles_Available'].tolist())   #sum of 3 or More Vehicles

p0 = (x1 / (x1+x2+x3+x4)) * 100   #proportion of No_Vehicle
p1 = (x2 / (x1+x2+x3+x4)) * 100   #proportion of 1_Vehicle
p2 = (x3 / (x1+x2+x3+x4)) * 100   #proportion of 1_Vehicle
p3 = (x4 / (x1+x2+x3+x4)) * 100   #proportion of 1_Vehicle

print(p0)
print(p1)
print(p2)
print(p3)
```

```python
# calculate DOWNTWON LA percentage

y1 = sum(df_sorted1['No_Vehicle_Available'].tolist())   #sum of No_Vehicle
y2 = sum(df_sorted1['1_Vehicle_Available'].tolist())   #sum of 1_Vehicle
y3 = sum(df_sorted1['2_Vehicle_Available'].tolist())   #sum of 2_Vehicle
y4 = sum(df_sorted1['3_or_More_Vehicles_Available'].tolist())   #sum of 3 or More Vehicles

q0 = (y1 / (y1+y2+y3+y4)) * 100   #proportion of No_Vehicle
q1 = (y2 / (y1+y2+y3+y4)) * 100   #proportion of 1_Vehicle
q2 = (y3 / (y1+y2+y3+y4)) * 100   #proportion of 1_Vehicle
q3 = (y4 / (y1+y2+y3+y4)) * 100   #proportion of 1_Vehicle

print(q0)
print(q1)
print(q2)
print(q3)
```

```python
p0 = round(p0, 2)
p1 = round(p1, 2)
p2 = round(p2, 2)
p3 = round(p3, 2)
q0 = round(q0, 2)
q1 = round(q1, 2)
q2 = round(q2, 2)
q3 = round(q3, 2)
```

## Plot a bar chart!

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 0.25
fig, ax = plt.subplots(figsize=(12, 8))

PACIFIC_PALISADES = [p0, p1, p2, p3]
DOWNTOWN_LA = [q0, q1, q2, q3]

br1 = np.arange(len(PACIFIC_PALISADES))
br2 = [x + barWidth for x in br1]

bars1 = ax.bar(br1, PACIFIC_PALISADES, color='red', width=barWidth,
               edgecolor='black', label='PACIFIC PALISADES')
bars2 = ax.bar(br2, DOWNTOWN_LA, color='yellow', width=barWidth,
               edgecolor='black', label='DOWNTOWN LA')

ax.set_ylabel('Number of Household %', fontweight='bold', fontsize=15)
ax.set_xticks([r + barWidth/2 for r in range(len(PACIFIC_PALISADES))])
ax.set_xticklabels(['No Vehicle Available', '1 Vehicle Available', '2 Vehicles Available', '3 or More Vehicles Available'])

ax.set_title('Household Vehicle Ownership Proportions of Community PACIFIC PALISADES and DOWNTOWN LA', fontweight='bold', fontsize=17, pad=20)

# adding number on top of each bar
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)

ax.legend()
plt.savefig('Vehicle Ownership Comparison')
plt.show()
```

In this chart, I compared the household vehicle owenership proportions between two communities: PACIFIC PALISADES and DOWNTOWN LA, which represent areas with low collision rate and high collision rate respectively. PACIFIC PALISADES exhibits a higher vehicle ownership compared to DOWNTOWN LA, where a significant portion of households do not possess vehicles. However, the number of collisions in DOWNTOWN LA experiences a substantially higher number of collisions, surpassing PACIFIC PALISADES by more than 10 times. This indicates that vehicle ownership alone does not correlate with collision rates.
